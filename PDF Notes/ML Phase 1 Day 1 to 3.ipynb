<h1>How to solve a ML Problem</h1>

- When it comes to devlopment of Machine Learning projects for a long time there was no standard process to develop the projects and as a direct consequence of that till date there are many projects which are still found to be very unorganized.
- This leads to another very big problem wherein if you are to recreate the same results it is a very hectic and tedious task.
- In order to cater the above issues in 1996, at a SIG workshop in  Brussels one of the very first frameworks was proposed which was called **<u>CRISP-DM</u>**
- CRISP-DM stands for **"CROSS INDUSTRY STANDARD PROCESS FOR DATA MINING"**
- After CRISP-DM in 2015 IBM came up with another framework called <u>**ASUM-DM**</u> which stands for <br><u>**"ANALYTICS SOLUTION UNIFIED METHOD FOR DATA MINING"**</u>
- In 2016 Microsoft came up with its framework called <u>**TDSP**</u> which stands for <u>**TEAM DATA SCIENCE PROCESS**</u> Lifecycle.
- Recently in 2021 another framework was proposed which was an extension on CRISP-DM and is called <u>**CRISP-ML(Q)**</u> which stands for **"CROSS INDUSTRY STANDARD PROCESS FOR DEVELOPMENT OF MACHINE LEARNING APPLICATIONS WITH QUALITY ASSURANCE METHODOLOGY"**

<h1> UNDERSTANDING THE PHASES OF CRISP-DM </h1>

<h2> 1. BUSINESS UNDERSTANDING</h2>

- <u>**1.1) Business Understanding Overview**</u>: Understanding the business reasons for the data mining efforts helps to ensure that the entire team is on the same page before any money is spent on any of the resources.
- <u>**1.2) Determining Business Objectives**</u>: Though not an easy task to do but once done, it helps in minimizing the risks by clarifying problems, goals and resources.
- <u>**1.3) Assessing the Situation**</u>: Once the goal has been determined, we need to understand the current whereabouts of the project which involves asking questions like:
  - What sort of data is available for analysis?
  - Do you have the personnel needed for the completion of the project?
  - What are the biggest risk factors involved?
  - What are the contigency plans for each of the risk?
- <u>**1.4) Determining Data Mining Goals**</u>: this point can be best explained with the help of an example.
For e.g. You have a telecomm business and you want to reduce the "churn" of your customers, this can be translated into the following goals:
  - Identifying high value customers based on recent purchase data.
  - Building a model using available customer data to predict the likelihood of churn for each customer.
  - Assigning each customer a rank based on both churn propensity and customer value.

These are some of the goals which if met can be used to reduce the churn of customers.
<br>
**NOTE: AN IMPORTANT OBSERVATION CAN BE MADE HERE THAT FOR AN EFFECTIVE DATA MINING BUSINESS AND TECHNOLOGY NEED TO GO HAND IN HAND.**
<br>
- <u>**1.5) Data Mining Goals**</u>: It is like maintaining a task list that needs to be solved using technical aspects.
For e.g.
  - Describe the type of data mining problem, such as clustering, prediction, or classification.
  - Document technical goals using specific units of time, such as predictions with a three-month validity.
  - If possible, provide actual numbers for desired outcomes, such as producing churn scores for 80% of existing customers.
- <u>**1.6) Data Mining Success Crieteria**</u>: Using the Data Mining Goals from the previous step we have to formulate benchmarks for success. Some of the pointers which can be considered are:
  - Describe the methods for model assessment (for example, accuracy, performance, etc.).
  - Define benchmarks for evaluating success. Provide specific numbers.
  - Define subjective measurements as best you can and determine the arbiter of success.
  - Consider whether the successful deployment of model results is part of data mining success.

<h2> 2. DATA UNDERSTANDING</h2>

Here we try to get a closer look at the data available to us in order to avoid unexpected problems which we can encounter in the next phase `DATA PREPARATION`.
<br>
In this phase we access the data and explore it using tables and graphics. This helps to determine the quality of data and describe the results of these steps in project documentation.<br>
- 2.1) <u>*Collecting Initial Data*</u>: Data comes from a variety of sources:
  - **Existing Data**: This encompasses a wide variety of data like: transactional data, survey data, web logs etc.
  - **Purchased Data**: It consists of supplemental data like demographics.
  - **Additional Data**: In case the data needs are not met with the above 2 then we may need to conduct surveys or begin additional tracking in order to meet the data requirements.

  Here we try to answer questions like:
    - Which attributes (columns) from the database seem most promising?
    - Which attributes seem irrelevant and can be excluded?
    - Is there enough data to draw generalizable conclusions or make accurate predictions?
    - Are there too many attributes for your modeling method of choice?
    - Are you merging various data sources? If so, are there areas that might pose a problem when merging?
    - Have you considered how missing values are handled in each of your data sources?

- 2.2) <u>*Describing data*</u>: While describing the data the following pointers can be utilized:
  - <u>`Amount of Data`</u>: There are a lot of important pointers here which need to be considered here like:
    - Data Size tradeoff: It simply means that for most of the modelling techniques larger data means better model but it comes at the cost processing time.
      - We can look into whether using a subset of the data is a possibility?
      - We need to look at the size statistics like the number of rows as well as the number of features available in the data.
  - <u>`Value Type`</u>: The type of data present in the various features of the data like `int`, `float` or `boolean`. Identifying this before hand can save a lot of time down the road in the project.
  - <u>`Coding Schemes`</u>: We try to identify if there are any conflicting coding schemes in data and try to mitigate them, for e.g. male and female in some data points can be represented as `m` and `f` respectively; alternatively it can also be represented using `1` and `2` respectively. Now this is a conflict in coding schemes which needs to be mitigated by selecting any of them for the entire data.

Also a lot common descriptions focus on the quantity and quality of the data
i.e. how much data is available and the condition of the data.
- 2.3) <u>*Exploring data*</u>: Here we try to explore the data with the help of tables, charts and other visualization tools in order to address the `data mining goal` constructed during the `business understanding phase`.
- 2.4) <u>*Verifying data quality*</u>: Maximum number of times the data that is provided is not perfect and can have issues like:
  - <u>`Missing Data`</u>: Blank or Null values
  - <u>`Data Errors`</u>: Typo errors made while entering the data
  - <u>`Measurement Errors`</u>: Data which was entered correctly but measured on a wrong scale or unit.
  - <u>`Coding Inconsistencies`</u>: When for the same data value there are multiple representations like in gender column we might have both `male` and `m`.
  - <u>`Bad Metadata`</u>: It is when we have a mismatch between the apparent meaning of the field and the meaning stated for the field in the problem statement


<h2> 3. DATA PREPARATION</h2>

- It is one of the most time consuming phases of Data Mining which generally takes somewhere around 70% of the project time. By ensuring that the previous phases are executed properly the overhead can be reduced but it still remains the most time consuming phase of all.<br> Depending on the project or task at hand it generally involves:
  - Merging Datasets
  - Selecting a sample subset of the data
  - Aggregating records
  - Derving new attributes
  - Sorting the data for modelling
  - Removing/Imputing missing values
  - Splitting data into train and test data

- 3.1) <u>**Data Selection**</u>: We select the relevant rows and columns for the data mining goals
- 3.2) <u>**Cleaning Data**</u>: We try to identify the problems with the data selected for analysis.

<u>**Data Problem**</u>  | <u>**Possible Solution**</u>
-------------------|------------------
**Missing Data**       | Exclude the rows or impute the missing values
**Data Errors**       | Exclude the characteristics or use logic to manually discover the errors and replace them
**Coding Inconsistencies**       | Decide on a single coding scheme, then convert and replace values
**Missing or Bad Metadata**       | Manually examine suspected fields and track down correct meaning

- 3.3) <u>**Constructing New Data**</u>: Here we try to generate new features in order to get better insights and relationship with target variable. It is generally done via the following 2 ways:
  - Derving attributes from the pre-existing attributes
  - Generating new rows
- 3.4) <u>**Integrating Data**</u>: There is always a possibility of scenarios where in the data can be present across multiple sources. Instead of working with all the sources separately we can merge all of them using a common key or feature.<br>
There are majorly 2 kinds of integrations:
  - **Merging**: It involves merging 2 datasets with similar records but different attributes.The data is merged using a common key or feature for each datapoint. The resultant data will have more columns/features.
  - **Appending**: It involves integrating 2 or more datasets with similar attributes but different records. The integration happens on the basis of a common field.<br>
- 3.5) <u>**Formatting Data**</u>: Here we try to check if the models that we want to train require the data in some particular format or in some order. While formatting the data try to consider these kind of questions:
  - Which models do you plan to use?
  - Do these models require a particular data format or order?<br>
  
Apply the necessary transformations as required to format the data


<h2> 4. MODELING</h2>

In data mining there are multiple ways to look at the problem and come up with multiple solutions. Since we can have multiple solutions, Modeling is conducted in multiple iterations. Generally several models are run using the default parameters, post which fine-tuning of the models is done.
<br>
4.1) <u>**Selecting Modeling Techniques**</u><br>
Determining the most appropriate model is based on:
- **Data Types Available For Mining**
- **Data Mining Goals**: Whether the goals is to just get the insights or to be able to perform predictions on the future datapoints.
- **Modelling Requirements**: Here we try to answer the questions like:
  - What size of data is required?
  - What type of data is required?
  - Do you need a model with easily presentable results.<br>

  Ensure that any assumptions that are made while modeling are noted down. For e.g. When using Naïve Bayes, mention the asumption of asuming all input features as independent.
<br>

4.2) <u>**Generating a Test Design**</u><br>
Before building the model, here we need to decide upon how do we plan to test the models results. There are two major parts on creating a detailed test design:
- Describing the criteria for "*goodness*" of a model
- Defining the data on which these criteria will be tested.<br>

Measuring a model's *goodness* can be done in several ways depending upon what the model is, For e.g. Decision Tree algorithm can have error rate, in case of unsupervised algorithms we can have ease of interpretation, deployment etc.
- Since the model building is an iterative process, it is imperative to know when to stop adjusting the parameters of the model or try other models. While creating a test design, try to answer the following:
  - What data will be used to test the models?
  - Have you partioned the data into train and test? Is so then in what ratio?
  - How to measure the success of Supervised Models?
  - How to measure the success of Unsupervised Models?
  - How many iterations for a model are allowed by changing the parameters before you move to next model?
<br>

4.3) <u>**Building The Models**</u><br>
- Build several models and compare the results before deploying or integrating them. Experiment with a number of different models before making final conclusions.
- In order to track your progress with a variety of models, be sure to keep notes on the settings and data used for each model.
- This will help you to discuss the results and retrace the steps if necessary.
- At the end of the model-building process, there should be three pieces of information to use in data mining decisions:
  - **Parameter settings** include the notes you take on parameters that produce the best results.
  - The actual **models** produced.
  - **Descriptions of model results**, including performance and data issues that occurred during the execution of the model and exploration of its results.

After training the model try to answer these questions as well for better understanding of the model:
- Can meaningful conclusions be drawn from this model?
- Are there new insights or unusual patterns revealed by the model?
- Were there execution problems for the model?
- How reasonable was the processing time?
- Did the model have difficulties with data quality issues, such as a high number of missing values?
- Were there any calculation inconsistencies that should be noted?
<br>

4.4) <u>**Assessing The Models**</u><br>
- At this point, you will have trained your models, so evaluate which of the following models are good enough to be put in production/ deployment etc.
- Once an assessment has been made, rank the models in order based on both objective (model accuracy) and subjective (ease of use or interpretation of results) criteria.<br>
Following questions can be used as reference for the same:
  - Using the evaluation charts, cross-validation charts to evaluate the results of your model.
  - Conduct a review of the results based on your understanding of the business problem. Consult data analysts or other experts who may have insight into the relevance of particular results.
  - Consider whether a model's results are easily deployable. Does your organization require that results be deployed over the Web or sent back to the data warehouse?
  - Analyze the impact of results on your success criteria. Do they meet the goals established during the business understanding phase?

If you were able to address the above issues successfully and believe that the current models meet your goals, it's time to move on to a more thorough evaluation of the models and a final deployment. Otherwise, take what you've learned and rerun the models with adjusted parameter settings.<br>
You have two options here:
  - Adjust the parameters of existing models.
  - Choose a different model to address your data mining problem.
  
In both cases, you'll be returning to the building models task and iterate until the results are successful.

<h2> 5. EVALUATION</h2>

At this stage we have successfuly build models which are technically correct and effective as per the <u>Data Mining Success Crieteria</u> as defined in earliers steps.<br>

Now we need to evaluate it as per the <u>Business Success Criteria</u> established in the initial steps. This evaluation will ensure that we can utilize the results obtained in previous steps.<br>

Data Mining is basically producing 2 results:
- <b>Final Model</b> from the previous step.
- <b>Findings</b>: It comprises of conclusions or inferences drawn from the model itself or the data mining process or both.

- <b><u>Evaluating the Result</b></u>:At this point we formalize our assessment of whether or not the project results meet the business success crieteria.
  - We do this by documenting the assessment of whether the data mining results meet the business success crieterai or not. This can be done easily if we try to answer the following questions in our report:
    - Are your results stated clearly and in a form that can be easily presented?
    - Are there particularly novel or unique findings that should be highlighted?
    - Can you rank the models and findings in order of their applicability to the business goals?
    - In general, how well do these results answer your organization's business goals?
    - What additional questions have your results raised? How might you phrase these questions in business terms?

- By drawing a line as shown above we can say that `any data point which lies below the line is a female and any data point above the line will be a male.`
- Now let's say we have 2 query points as:
  - Q1 : (140,50)
  - Q2 : (190,90)

and we want to find the gender for these points.

````python
# Plotting the query points
plt.plot([145,185],[80,60], linestyle="--")
sns.set(style="darkgrid")
sns.scatterplot(df,x=df["Height"],y=df["weight"],hue=df["gender"],palette="pastel")
plt.plot([140],[50],'ro') # Red query point Q1
plt.plot([190],[90],'ro') # Red query point Q2
plt.text(138,52,"Q1")
plt.text(189,87,"Q2")
````

- Now looking at the query points in the graph with respect to the line drawn:
  - `Q1 is Female`
  - `Q2 is Male`
- There is another approach to look at it, if we see the majority of the neighbours of each query point, then based on the similarity of neighbours we can again classify the query points as:
  - `Q1 -> Female` as all nearby points are female
  - `Q2 -> Male` as nearby points are male

So now lets try to compile the points for better understanding:

1.   `Method 1`: <b>KNN way</b> -> based upon the types/similarity and number of neighbours we classify the query point
2.   `Method 2`: <b>Boundary Way</b> -> Using a line to make one side as male and other side as female. This is used by Logistic Regression, SVM etc.
3. `Method 3`: <b>Probabilistic Way</b> -> This is used by Naïve Bayes.
4. `Method 4`: <b>Rule Based Way</b> -> Decision Tree automatically comes up with these rules.
5. `Method 5`: <b>Using Ensembles</b> -> Combination of algorithm to either comes up with rules or boundaries to classify the points
6. `Method 6`: <b>Deep Learning</b> -> A.N.N.(Artificial Neural Network) is a complex architecture which tries to classify the points



Now since we know about various M.L. Models which can help us solve our tasks, lets have a brief look at the ways to evaluate if the model is performing properly or not:
- For `Regression` Tasks:
  - Mean Squared Error (MSE)
  - Root Mean Squared Error (RMSE)
  - Mean Absolute Error (MAE)
  - Coefficient of Determination (R2 Score)
  - Adjusted R2 Score
- For `Classification` Tasks:
  - Accuracy
  - Confusion Matrix
  - Precision and Recall
  - F1-Score
  - ROC-AUC (Reciever Operating Characteristic and Area Under the Curve)
  - Log Loss

<u>**NOTE**</u>: M.L. Algorithms are not always right and have some small amount of `Irreducible Error` in them.

In order to get started with any M.L. problem statement, try to answer the following questions:
- <b>Is historical data available?</b>
  - `Yes` -> Proceed forward
  - `No` -> Do data collection
- <b>What is the target variable?</b>
- <b>Is the target variable given in historical data?</b>
  - `Yes` -> Supervised Learning
  - `No` -> Unsupervised Learning
- <b>Is the target variable discrete or continuous?</b>
  - `Discrete` -> Classification Task
  - `Continuous` -> Regression Task
