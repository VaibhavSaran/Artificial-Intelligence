```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
import numpy as np
```

```python
url = "https://www.makaan.com/hyderabad-residential-property/rent-property-in-hyderabad-city?_=1699296035195"
price = []
page = requests.get(url)
soup = BeautifulSoup(page.text)
for i in soup.find_all("td",class_="price"):
        text = i.text
        # print(text)
        text = text.replace(',','')
        text = text.strip()
        regex = re.findall("^\d+\.?\d?",text)
        if regex:
            price.append(regex[0]) 
        else:
            price.append(np.nan)
print(price)
```

- The basic idea is that perform scraping for every URL and then go to the next one

```python
bhk = []
locality = []
price = []
area = []
status = []
flat_type = []
for i in range(1,40):
    url = f"https://www.makaan.com/hyderabad-residential-property/rent-property-in-hyderabad-city?_=1699296035195&page={i}"
    page = requests.get(url)
    soup = BeautifulSoup(page.text)
    # For BHK
    for i in soup.find_all('div',class_='title-line'):
        text = i.text
        regex = re.findall("\d\sBHK",text)
        if regex:
            bhk.append(regex[0]) 
        else:
            bhk.append(np.nan)

    # For Flat type
    for i in soup.find_all('div',class_='title-line'):
        text = i.text
        regex = re.findall("\d+\s(?:BHK)?\s?(\w+)",text)
        if regex:
            flat_type.append(regex[0]) 
        else:
            flat_type.append(np.nan)
    
    # For Locality
    for i in soup.find_all('div',class_="locWrap"):
        text = i.text
        regex = re.findall("(^[a-zA-z]+\s?\w+)",text)
        if regex: 
            locality.append(regex[0]) 
        else:
            locality.append(np.nan)

    # For Price
    for i in soup.find_all("td",class_="price"):
        text = i.text
        text = text.replace(',','')
        text = text.strip()
        regex = re.findall("^\d+\.?\d?\s?\w+",text)
        if regex:
            price.append(regex[0]) 
        else:
            price.append(np.nan)

    # For Square Foot Area of estate
    for i in soup.find_all("td",class_="size"):
        text = i.text
        regex = re.findall("\d+",text)
        if regex:
            area.append(regex[0])
        else:
            area.append(np.nan)
    
    #For Finding the furnishing status of the property
    for i in soup.find_all("tr",class_='hcol w44'):
        text = i.text
        if text:
            text = text[:len(text)-6]
            status.append(text)
        else:
            status.append(np.nan)

# Task Scrape the directions of the flat, number of washrooms and if a deposit is required or not
print(len(bhk),len(flat_type),len(locality),len(area),len(status),len(price))
```
