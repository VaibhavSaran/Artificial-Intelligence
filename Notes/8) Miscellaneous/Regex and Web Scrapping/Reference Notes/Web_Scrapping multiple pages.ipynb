# WebScraping

##### For collecting large amount of data from a website for analysis we use webscrapping.
##### It is a automated process for collecting the data and storing it in a structured format.
##### Access the HTML of the webpage and extract useful information from it.

# Libraries Used
1. requests : to extract HTML code from the website
2. BeautifulSoup : format and gets the components from the HTML code

### Requests module 
- In order to get data we make a request to webserver.

- The server then replies with our data.

### Steps

- Identify URL
- Inspect HTML code
- Find HTML tag and element that you want to extract
- Write code to scrape the data

# Status code

- 200 - Success

- 404 - Page not found

- 403 - Client side error

- 500 - server side error

- 503 - server side error

 - For 403 error , use the following syntax(header).

    requests.get(url,headers = {'User-Agent':'Mozilla/5.0'})

## For multiple Pages

```python
import pandas as pd
import numpy as np
```

```python
prices = []
ram = []
ratings=[]
brand = []

for i in range(1,5):
    url =f'https://www.flipkart.com/search?q=mobiles&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={i}'
    
    response = requests.get(url)
    soup = BeautifulSoup(response.text)
    
    for i in soup.find_all('div', class_='_2kHMtA'):
        
        price=i.find('div', class_ = '_30jeq3 _1_WHN1')
        prices.append(price.text)
        
        ram_ = i.find('div', class_='_4rR01T').text
        regex  = re.findall('\d+\sGB', ram_)
        if regex:
            ram.append(regex[0])
        else:
            ram.append(np.nan)
            
        rating = i.find('div', class_= '_3LWZlK')
        if rating:
            ratings.append(rating.text)
        else:
            ratings.append(np.nan)
            
        brand_ =i.find('div', class_='_4rR01T').text
        regex = re.findall(r'^(\w+)', brand_)
        if regex:
            brand.append(regex[0])
        else:
            brand.append(np.nan)
```

```python
ram
```

```python
prices
```

```python
df = pd.DataFrame({'prices':prices,
                  'ram':ram,
                  'ratings':ratings,
                  'brand':brand})
```
