```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
import numpy as np
```

```python
url="https://www.flipkart.com/search?q=mobiles&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&p%5B%5D=facets.price_range.from%3D10000&p%5B%5D=facets.price_range.to%3DMax&page=4"
```

```python
for i in range(1,4):
    print(i)
```

```python
brand=[]
color=[]
storage=[]
rating=[]
price=[]
battery=[]
ss=[]
ram=[]
for i in range(1,21):
    url=f"https://www.flipkart.com/search?q=mobiles&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&p%5B%5D=facets.price_range.from%3D10000&p%5B%5D=facets.price_range.to%3DMax&page={i}"
    
    page=requests.get(url)
    
    soup=BeautifulSoup(page.text)
    
    for i in soup.find_all("div",class_="_4rR01T"):
        #print(i.text)
        regex=re.findall("^[a-zA-Z]+",i.text)
        #print(regex)
        if regex:
            brand.append(regex[0])
        else:
            brand.append(np.nan)
            
    for i in soup.find_all("div",class_="_4rR01T"):
        print(i.text)
        regex=re.findall("(\d+)\sGB",i.text)
        print(regex)
        if regex:
            storage.append(regex[0])
        else:
            storage.append(np.nan)
    
    for i in soup.find_all("div",class_="_4rR01T"):
        print(i.text)
        regex=re.findall("\((\w+\s?\w+),",i.text)
        print(regex)
        if regex:
            color.append(regex[0])
        else:
            color.append(np.nan)
    
    container=soup.find_all("div",class_="_2kHMtA")
    for i in container:
        a=i.find("div",class_="_3LWZlK")

        if a:
            rating.append(a.text)
        else:
            rating.append(np.nan)
            
    for i in container:
        b=i.find("div",class_="_30jeq3 _1_WHN1")
        if b:
            
            price.append(b.text)
        else:
            price.append(np.nan)
    
    for i in container:
        c=i.find("div",class_="col col-7-12")
        print(c.text)
        regex=re.findall("(\d+)\smAh",c.text)
        if regex:
            battery.append(regex[0])
        else:
            battery.append(np.nan)
    
    for i in container:
        c=i.find("div",class_="col col-7-12")
        print(c.text)
        regex=re.findall("(\d+\.?\d+)\sinch",c.text)
        if regex:
            ss.append(regex[0])
        else:
            ss.append(np.nan)
    for i in container:
        c=i.find("div",class_="col col-7-12")
        print(c.text)
        regex=re.findall("(\d+)\sGB\sRAM",c.text)
        if regex:
            ram.append(regex[0])
        else:
            ram.append(np.nan)
```

# WebDriver
> The webdriver in Selenium is an automation framework used to carry out testing in the web in multiple browsers.

> As a test script is executed, a HTTP request is generated for every command in Selenium and passed to the browser driver. The browser driver takes the help of a HTTP server for obtaining the HTTP request.

The types of Browser drivers are:

- ChromeDriver

- FirefoxDriver.

- InternetExplorerDriver

- SafariDriver
